```python
hubble = 0.70
```

```python
particle_data = loadtxt("../data/halo_85000034071_particle_ID.csv", skiprows=16, delimiter=',')
```

```python
x_p = particle_data[:,2]
y_p = particle_data[:,3]
z_p = particle_data[:,4]
```

```python
scatter(x_p, y_p, s=0.1)
```

```python
mean_x_p = mean(x_p)
mean_y_p = mean(y_p)
mean_z_p = mean(z_p)
std_x_p = std(x_p)
std_y_p = std(y_p)
std_z_p = std(z_p)
avg_std = (std_x_p + std_y_p + std_z_p)/3.0
```

```python
print std_x_p, std_y_p, std_z_p, avg_std
```

```python
#select data cube within cube_sigma times the standard deviation
cube_sigma = 1.0
cube_index = where((fabs(x_p-mean_x_p)<(cube_sigma*avg_std))&(fabs(y_p-mean_y_p)<(cube_sigma*avg_std))&(fabs(z_p-mean_z_p)<(cube_sigma*avg_std)))
cube_index = cube_index[0]
```

```python
scatter(x_p[cube_index], y_p[cube_index], s=0.1)
```

```python
n_particles_in = size(cube_index)
n_side = int(sqrt(n_particles_in))*4
delta_field = 2.0*cube_sigma*avg_std/n_side
print n_side, delta_field
```

```python
kappa_field = zeros((n_side, n_side))
for index in cube_index:
    x = x_p[index] - (mean_x_p -  cube_sigma*avg_std)
    y = y_p[index] - (mean_y_p -  cube_sigma*avg_std)
    
    i_field = int_(x/delta_field)
    j_field = int_(y/delta_field)
    
    b_x = (x - i_field*delta_field)/delta_field
    b_y = (y - j_field*delta_field)/delta_field
    
    peso_x = 1.0-b_x
    peso_y = 1.0-b_y
    peso_x_plus = 1.0 - peso_x
    peso_y_plus = 1.0 - peso_y
    
    masa = 1.0
    
    kappa_field[i_field, j_field] = kappa_field[i_field, j_field] + masa
#    if(i_field+1<n_side):
#        kappa_field[i_field+1, j_field] = kappa_field[i_field, j_field] + masa*peso_x_plus*peso_y
#    if(j_field+1<n_side):
#        kappa_field[i_field, j_field+1] = kappa_field[i_field, j_field] + masa*peso_x*peso_y_plus
#    if((i_field+1<n_side)&(j_field+1<n_side)):
#        kappa_field[i_field+1, j_field+1] = kappa_field[i_field, j_field] + masa*peso_x_plus*peso_y_plus

kappa_field = kappa_field/((delta_field/hubble)**2)
```
